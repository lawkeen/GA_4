```python
import numpy as np
import pandas as pd
import pygad
```

```python
df = pd.read_csv('13.txt', delimiter=' ', engine='python')
df
```

```python
wv = df.columns.tolist()
W = int(wv[0]) 
V = int(wv[1]) 
df.columns = ['v', 'c']
df['w'] = df.index
data = df[['w','v','c']].values.tolist()
```

```python
#Функция, вычисляющая ценность решения
def fitness_func(solution, solution_idx):
    w, v, c = 0, 0, 0
    for (array, item) in zip(solution, data):
        if array > 0:
            w += item[0]
            v += item[1]
            c += item[2]
    if w > W or v > V:
        c = 0
    return c
```

```python
#Создаем экземпляр класса pygad.GA 
ga_instance = pygad.GA(num_generations=200, 
                       num_parents_mating=2, 
                       fitness_func=fitness_func, 
                       sol_per_pop=10, 
                       num_genes=len(data),
                       init_range_low=-1, 
                       init_range_high=1 
                       )
```

```python
#Запускаем генетический алгоритм
ga_instance.run()
solution, solution_fitness, _ = ga_instance.best_solution()
```

```python
result = []
for i in range(0, len(data)):
    if solution[i] > 0:
        result.append(data[i])
result_data = pd.DataFrame.from_records(result, columns=['w','v','c'])
```

```python
print(solution_fitness)
print(result_data)
w, v, c = 0, 0, 0
for i in range(0, len(result_data)):
        w += result_data.at[i,'w']
        v += result_data.at[i,'v']
        c += result_data.at[i,'c']
print("Вес:", w)
print("Объем:", round(v,1))
print("Ценность:", c)
```

```python
file=open('GA_4_1.txt','w', encoding='utf-8')
file.write(result_data.to_string())
file.write("\nВес: " + str(w))
file.write("\nОбъем: " + str(round(v,1)))
file.write("\nЦенность: " + str(c))
file.close()
```
